```python
import os
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

gpkg = "data/Penang_LandUse_2024_SPU_SPT.gpkg"
layer = "Penang_LandUse_2024_SPU_SPT"
output_path = "image/Penang_LandUse_2024_Coloured_Map.png"

os.makedirs("image", exist_ok=True)

gdf = gpd.read_file(gpkg, layer=layer)

colors = {
    "Pertanian": "#7fc97f",
    "Perumahan": "#ffff99",
    "Tanah Kosong": "#fdae61",
    "Pengangkutan": "#80b1d3",
    "Industri": "#bebada",
    "Badan Air": "#1f78b4",
    "Hutan": "#006d2c",
    "Institusi dan Kemudahan Masyarakat": "#fb9a99",
    "Komersial": "#e41a1c",
    "Tanah Lapang dan Rekreasi": "#b2df8a",
    "Infrastruktur dan Utiliti": "#cab2d6",
    "Pantai": "#fdbf6f",
    "Tanah Lapang": "#dddddd",
}

gdf["plot_color"] = gdf["gtn1"].map(colors).fillna("#cccccc")

# Dissolve 212,708 polygons into land-use classes for faster plotting
gdf_diss = gdf[["gtn1", "plot_color", "geometry"]].dissolve(by="gtn1")

# Simplify geometry to reduce rendering load
gdf_diss["geometry"] = gdf_diss.geometry.simplify(
    tolerance=5,
    preserve_topology=True
)

fig, ax = plt.subplots(figsize=(14, 12))

gdf_diss.plot(
    ax=ax,
    color=gdf_diss["plot_color"],
    edgecolor="black",
    linewidth=0.2
)

legend_items = [
    Patch(facecolor=color, edgecolor="black", label=label)
    for label, color in colors.items()
    if label in set(gdf["gtn1"])
]

ax.legend(
    handles=legend_items,
    title="Land Use Class",
    loc="center left",
    bbox_to_anchor=(1.02, 0.5),
    frameon=True,
    fontsize=9,
    title_fontsize=10,
    borderaxespad=0.0
)

ax.set_axis_off()

plt.subplots_adjust(right=0.78)

plt.savefig(
    output_path,
    dpi=200,
    bbox_inches="tight"
)

plt.close()

print("Saved:", os.path.abspath(output_path))
```
